```python
import pandas as pd
import numpy as np
from google.colab import files
import io

def clean_dataframe(df):
    """
    Cleans the DataFrame by removing rows with 0 or N/A values and
    dropping columns with variance less than 10.
    """
    # Create a copy to keep the original DataFrame safe
    df_cleaned = df.copy()

    # --- 1. Clean the data ---
    # Replace all 0 values with NaN (Not a Number) to handle them together
    df_cleaned.replace(0, np.nan, inplace=True)

    # Drop all rows that contain at least one NaN value
    df_cleaned.dropna(inplace=True)

    # --- 2. Remove columns with low variance ---
    # Select only the columns that contain numbers
    numeric_cols = df_cleaned.select_dtypes(include=np.number).columns

    # Check if there are any numeric columns to process
    if not numeric_cols.empty:
        # Calculate the variance for each numeric column
        variances = df_cleaned[numeric_cols].var()

        # Find the columns where the variance is less than 10
        cols_to_drop = variances[variances < 10].index

        # Drop these columns from your dataframe
        df_cleaned.drop(columns=cols_to_drop, inplace=True)

        print("\nColumns dropped because their variance was less than 10:")
        if len(cols_to_drop) > 0:
            print(list(cols_to_drop))
        else:
            print("No columns had a variance less than 10.")

    else:
        print("\nNo numeric columns found to calculate variance.")

    return df_cleaned


# --- Main script execution ---
print("Please choose your CSV file to upload and clean.")

# Trigger the file upload dialog
uploaded = files.upload()

# Check if a file was uploaded
if not uploaded:
    print("\nNo file was selected. Please run the cell again.")
else:
    # Get the filename of the uploaded file
    file_name = next(iter(uploaded))

    print(f"\nSuccessfully uploaded '{file_name}'")

    try:
        # Read the uploaded file's content into a pandas DataFrame
        # uploaded[file_name] contains the raw bytes of the file
        df = pd.read_csv(io.BytesIO(uploaded[file_name]))

        print(f"Original number of rows and columns: {df.shape}")

        # Clean the DataFrame using the function defined above
        cleaned_df = clean_dataframe(df)

        print(f"Shape after cleaning: {cleaned_df.shape}")

        # --- Save and download the cleaned data ---
        # Convert the cleaned dataframe to a CSV string
        cleaned_csv = cleaned_df.to_csv(index=False)

        # Define the name for the new cleaned file
        output_filename = 'cleaned_file.csv'

        # Download the cleaned file to your local machine
        files.download(output_filename)
        print(f"\nDownload of '{output_filename}' should start automatically.")

    except Exception as e:
        print(f"\nAn error occurred while processing the file: {e}")
        print("Please ensure you uploaded a valid CSV file.")
```

```python
import pandas as pd
import numpy as np
from google.colab import files
import io

def clean_dataframe(df):
    """
    Cleans the DataFrame with more intelligent options.
    """
    # Create a copy to keep the original DataFrame safe
    df_cleaned = df.copy()

    # --- 1. Handle 0s and N/A values ---
    print("\n--- Step 1: Handling 0s and N/As ---")

    # Replace all 0 values with NaN (Not a Number) to handle them together
    df_cleaned.replace(0, np.nan, inplace=True)

    # --- CHOOSE YOUR CLEANING METHOD ---
    # Option 1: 'fill_mean' -> Fills missing values with the average of the column. (RECOMMENDED)
    # Option 2: 'drop_rows' -> Deletes any row with at least one missing value. (CAUTION: This made your last file empty)

    cleaning_method = 'fill_mean' # <--- You can change this to 'drop_rows' if you want

    if cleaning_method == 'fill_mean':
        print("Cleaning method: Filling missing values with their column's mean.")
        # For each numeric column, fill NaN with the column's average
        for col in df_cleaned.select_dtypes(include=np.number).columns:
            df_cleaned[col].fillna(df_cleaned[col].mean(), inplace=True)
    elif cleaning_method == 'drop_rows':
        print("Cleaning method: Dropping all rows with any missing values.")
        df_cleaned.dropna(inplace=True)

    print("Step 1 complete.")

    # --- 2. Remove columns with low variance ---
    print("\n--- Step 2: Removing Low-Variance Columns ---")

    # Select only the columns that contain numbers
    numeric_cols = df_cleaned.select_dtypes(include=np.number).columns

    if not numeric_cols.empty:
        variances = df_cleaned[numeric_cols].var()
        cols_to_drop = variances[variances < 10].index
        df_cleaned.drop(columns=cols_to_drop, inplace=True)

        print("Columns dropped because their variance was less than 10:")
        if len(cols_to_drop) > 0:
            print(list(cols_to_drop))
        else:
            print("No columns had a variance less than 10.")
    else:
        print("No numeric columns found to calculate variance.")

    print("Step 2 complete.")
    return df_cleaned


# --- Main script execution ---
print("Please choose your CSV file to upload and clean.")
uploaded = files.upload()

if not uploaded:
    print("\nNo file was selected. Please run the cell again.")
else:
    file_name = next(iter(uploaded))
    print(f"\nSuccessfully uploaded '{file_name}'")

    try:
        df = pd.read_csv(io.BytesIO(uploaded[file_name]))
        print(f"Original shape: {df.shape}")

        cleaned_df = clean_dataframe(df)
        print(f"\nFinal shape after cleaning: {cleaned_df.shape}")

        # --- Save and download the cleaned data (FIXED) ---
        output_filename = 'cleaned_file.csv'

        # **FIX:** First, save the dataframe to a file in Colab's storage
        cleaned_df.to_csv(output_filename, index=False)

        # **FIX:** Then, trigger the download for that existing file
        files.download(output_filename)
        print(f"\nYour cleaned file '{output_filename}' should be downloading now.")

    except Exception as e:
        print(f"\nAn error occurred while processing the file: {e}")
```

```python
import pandas as pd
import numpy as np
from google.colab import files
import io

# --- Main script execution ---
print("Please choose the CSV file from which to remove low-variance columns.")

# Trigger the file upload dialog
uploaded = files.upload()

if not uploaded:
    print("\nNo file was selected. Please run the cell again.")
else:
    # Get the filename of the uploaded file
    file_name = next(iter(uploaded))

    print(f"\nSuccessfully uploaded '{file_name}'")

    try:
        # Read the uploaded file into a pandas DataFrame
        df = pd.read_csv(io.BytesIO(uploaded[file_name]))

        print(f"\nOriginal shape: {df.shape}")

        # --- 1. Identify and drop low-variance columns ---

        # Select only the columns that contain numbers, as variance can only be calculated on them.
        numeric_cols = df.select_dtypes(include=np.number)

        if numeric_cols.empty:
            print("No numeric columns found in the file to calculate variance.")
        else:
            # Calculate the variance for each numeric column
            variances = numeric_cols.var()

            # Identify columns where the variance is less than 10
            cols_to_drop = variances[variances < 10].index

            # Drop these columns from the original dataframe
            df.drop(columns=cols_to_drop, inplace=True)

            # --- 2. Display summary ---
            print("\n--- Cleaning Summary ---")
            if len(cols_to_drop) > 0:
                print(f"Dropped the following {len(cols_to_drop)} columns with variance < 10:")
                # Print the list of dropped columns
                for col_name in cols_to_drop:
                    print(f"- {col_name} (Variance: {variances[col_name]:.4f})")
            else:
                print("No columns had a variance less than 10.")

            print(f"\nFinal shape after cleaning: {df.shape}")

            # --- 3. Save and download the cleaned data ---
            output_filename = 'variance_cleaned_file.csv'
            df.to_csv(output_filename, index=False)
            files.download(output_filename)
            print(f"\nYour cleaned file '{output_filename}' should be downloading now.")

    except Exception as e:
        print(f"\nAn error occurred while processing the file: {e}")
        print("Please ensure you uploaded a valid CSV file.")
```

```python
import pandas as pd
import numpy as np
from google.colab import files
import io

# --- Main script execution ---
print("Please choose the CSV file to process.")

# Trigger the file upload dialog in Colab
uploaded = files.upload()

if not uploaded:
    print("\nNo file was selected. Please run the cell again.")
else:
    # Get the filename of the single uploaded file
    file_name = next(iter(uploaded))

    print(f"\nSuccessfully uploaded '{file_name}'")

    try:
        # Read the uploaded CSV data into a pandas DataFrame
        df = pd.read_csv(io.BytesIO(uploaded[file_name]))

        print(f"\nOriginal shape: {df.shape}")

        # --- 1. Define the column to exclude from the variance check ---
        column_to_exclude = 'Exp_Value: -Log10(mol/kg)'

        # --- 2. Identify and drop low-variance columns, respecting the exclusion ---

        # Select only the columns that contain numbers
        numeric_cols_df = df.select_dtypes(include=np.number)

        if numeric_cols_df.empty:
            print("No numeric columns found in the file to calculate variance.")
        else:
            # Create a list of numeric column names to check
            cols_to_check = list(numeric_cols_df.columns)

            # Check if the exclusion column exists and remove it from our check-list
            if column_to_exclude in cols_to_check:
                cols_to_check.remove(column_to_exclude)
                print(f"\nSuccessfully excluded the column '{column_to_exclude}' from the variance check.")
            else:
                print(f"\nWarning: The specified exclusion column '{column_to_exclude}' was not found in the numeric columns.")

            # Calculate the variance ONLY on the columns we need to check
            variances = numeric_cols_df[cols_to_check].var()

            # Identify columns from our check-list where the variance is less than 10
            cols_to_drop = variances[variances < 10].index

            # Drop these columns from the main dataframe
            df.drop(columns=cols_to_drop, inplace=True)

            # --- 3. Display summary ---
            print("\n--- Cleaning Summary ---")
            if len(cols_to_drop) > 0:
                print(f"Dropped {len(cols_to_drop)} columns with variance < 10:")
                for col_name in cols_to_drop:
                    # Display the variance rounded to 4 decimal places
                    print(f"- {col_name} (Variance: {variances[col_name]:.4f})")
            else:
                print("No columns (besides the excluded one) had a variance less than 10.")

            print(f"\nFinal shape after cleaning: {df.shape}")

            # --- 4. Save and download the cleaned data ---
            output_filename = 'variance_cleaned_final.csv'
            df.to_csv(output_filename, index=False)
            files.download(output_filename)
            print(f"\nYour cleaned file '{output_filename}' should be downloading now.")

    except Exception as e:
        print(f"\nAn error occurred while processing the file: {e}")
        print("Please ensure you uploaded a valid CSV file.")
```
